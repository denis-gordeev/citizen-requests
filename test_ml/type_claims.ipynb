```python
class MultyFuzzyClass:
    classes = []
    probes = []
    tokens = []

    def __init__(self, classes, level=80, length=5):
        self.russian_stopwords = stopwords.words('russian')
        self.classes = classes
        self.level = level
        self.length = length

    def fit(self, text):
        self.probes = []
        self.tokens = [word for word in self.preprocess_text(text) if len(word) >= self.length]

    def transform(self):
        for cl in self.classes:
            self.probes.append(process.extractOne(cl, self.tokens))
        return

    def preprocess_text(self, text):
        text = re.sub('[%s]' % re.escape(punctuation), ' ', text)  # замена пунктуации на пробелы
        tokens = word_tokenize(text.lower(), language='russian')
        tokens = [token for token in tokens if token not in self.russian_stopwords \
                  and token != " " \
                  and token.strip() not in punctuation]
        return tokens
```
